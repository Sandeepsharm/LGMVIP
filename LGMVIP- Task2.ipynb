```python
# 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

```

### READ THE IMAGE IN RGB FORMAT




```python
# READ THE IMAGE IN RGB FORMAT
img = cv2.imread('image1.jpg')
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),cmap='gray')
plt.title("Original Image")
plt.show()
```

### CREATING A NEW IMAGE BY CONVERTING THE ORIGINAL IMAGE TO GREYSCALE


```python
# CREATING A NEW IMAGE BY CONVERTING THE ORIGINAL IMAGE TO GREYSCALE

gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_image,cmap='gray')
plt.title("Greyscale Image")
plt.show()
```

### INVERTING IMAGE


```python
#INVERTING IMAGE OF IRON MAN

invert_image = 255 - gray_image
plt.figure(figsize=(10,10))
plt.imshow(invert_image,cmap='gray')
plt.title("Smoothen Image")
plt.show()
```

### BLUR IMAGE BY USING THE GAUSSIAN FUNCTION

```python
#BLUR IMAGE OF BY USING THE GAUSSIAN FUNCTION
blurred = cv2.GaussianBlur(invert_image, (21, 21), 0)
plt.figure(figsize=(10,10))
plt.imshow(blurred,cmap='gray')
plt.title("Blurred Image")
plt.show()
```

### INVERT BLURRED IMAGE AS ORIGINAL IMAGE


```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),cmap='gray')
plt.title("Oroginal Image")
plt.show()
```

### ORIGINAL IMAGE TO PENCIL SKETCH

```python

plt.figure(figsize=(10,10))
plt.imshow(pencil_sketch,cmap='gray')
plt.title("Pencil Scratch Image")
plt.show()
```
